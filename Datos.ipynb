INICIO DE DIAGRAMAS PROYECTO BOOTCAMP

IMPORTS

```python
# Importar librerias necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

RECOPILACION DE DATOS

```python
# Codigo inicial y carga de datos

import os
file_path = 'Electricity generation by source - Colombia.csv'
if not os.path.exists(file_path):
	print(f"The file '{file_path}' does not exist in the current directory: {os.getcwd()}")
	print(f"Absolute path to check: {os.path.abspath(file_path)}")
	raise FileNotFoundError(f"The file '{file_path}' does not exist. Please ensure it is in the correct directory.")

Data = pd.read_csv(file_path)
Data.head()
```

```python
Data = Data.drop(columns=['Coal', 'Oil', 'Natural gas'], errors='ignore')
# Eliminar columnas innecesarias
```

IMPORTAR DATOS PARA DIAGRAMAS

```python
# Seleccionar columnas de interes para el analisis y renombre de columnas
Data['Country'] = 'Colombia'

Grupos = Data.groupby('Country')
EnergyCol = Grupos.get_group('Colombia')
EnergyCol
```

INICIO DE DIAGRAMAS

DIAGRAMA DE LINEAS

```python
# codigo base para graficar el primer diagrama
DataAll = EnergyCol[['Biofuels','Hydro','Solar PV','Wind','Other sources']]
plt.plot(EnergyCol['year'], DataAll, label=DataAll.columns)
plt.title('Generacion Electrica por source - Colombia')
plt.xlabel('Años')
plt.ylabel('Generacion electrica (GWh)')
plt.legend(DataAll.columns, loc='upper left')
plt.grid()
plt.show()
```

```python
# Separacion y datos de todas las energias
Total_Energia = np.sum(EnergyCol.select_dtypes(include='number'))
Total_Energia = Total_Energia.iloc[1:]
Total_Energia
```

DIAGRAMA DE BARRAS

```python
plt.figure(figsize=(10, 5))
plt.bar(Total_Energia.index, Total_Energia.values , color=['aquamarine', 'green', 'grey', 'blue', 'orange'])
plt.title('Generacion Electrica por source - Colombia')
plt.xlabel('Años')
plt.ylabel('Generacion de electricidad (GWh)')
plt.xticks(rotation=45)
```

DIAGRAMA DE BARRAS INDIVIDUAL

```python
# Primer Diagrama sobre la generacion de energia hidroelectrica
plt.hist(DataAll['Hydro'], bins=7, color='blue', alpha=0.7, label='Hydro', edgecolor='black')
plt.title('Generacion Electrica por source - Colombia')
plt.xlabel('Generacion de energia Hidroelectrica (GWh)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# segundo diagrama sobre el Biocombustible
plt.hist(DataAll['Biofuels'], bins=7, color='Green', alpha=0.7, label='Biofuels', edgecolor='black')
plt.title('Generacion Electrica por source - Colombia')
plt.xlabel('Generacion de biocombustibles (GWh)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Tercer diagrama sobre la energia solar 
plt.hist(DataAll['Solar PV'], bins=7, color='Yellow', alpha=0.7, label='Solar PV', edgecolor='black')
plt.title('Generacion Electrica por source - Colombia')
plt.xlabel('Genereacion solar PV (GWh)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Cuarto diagrama sobre fuentes de energia eolicas
plt.hist(DataAll['Wind'], bins=7, color='Grey', alpha=0.7, label='Wind', edgecolor='black')
plt.title('Generacion Electrica por source - Colombia')
plt.xlabel('Generación eólica (GWh)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Quinto diagrama sobre otras fuentes de energia
plt.hist(DataAll['Other sources'], bins=7, color='Black', alpha=0.7, label='Other sources', edgecolor='black')
plt.title('Generacion Electrica por source - Colombia')
plt.xlabel('Generacion de otras fuentes (GWh)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Diagrma de barras individuales y agrupados en un solo grafico
# Diagrama de hidroelectricidad
fig, ax = plt.subplots(2,3, figsize=(15,15), dpi=80, facecolor='w', edgecolor='k')
ax[0,0].hist(DataAll['Hydro'], bins=7, color='blue', alpha=0.7, label='Hydro', edgecolor='black')
ax[0,0].set_title('Generacion Electrica por source - Colombia')
ax[0,0].set_xlabel('generacion de energia Hidroelectrica (GWh)')
ax[0,0].set_ylabel('Frecuencia')
ax[0,0].grid()
# Diagrama de biocombustibles
ax[0,1].hist(DataAll['Biofuels'], bins=7, color='Green', alpha=0.7, label='Biofuels', edgecolor='black')
ax[0,1].set_title('Generacion Electrica por source - Colombia')
ax[0,1].set_xlabel('Generacion de biocombustibles (GWh)')
ax[0,1].set_ylabel('Frecuencia')
ax[0,1].grid()
# Diagrama de energia solar
ax[0,2].hist(DataAll['Solar PV'], bins=7, color='Yellow', alpha=0.7, label='Solar PV', edgecolor='black')
ax[0,2].set_title('Generacion Electrica por source - Colombia')
ax[0,2].set_xlabel('Generacion solar PV (GWh)')
ax[0,2].set_ylabel('Frecuencia')
ax[0,2].grid()
# Diagrama de energia eolica
ax[1,0].hist(DataAll['Wind'], bins=7, color='Grey', alpha=0.7, label='Wind', edgecolor='black')
ax[1,0].set_title('Generacion Electrica por source - Colombia')
ax[1,0].set_xlabel('Generación eólica (GWh)')
ax[1,0].set_ylabel('Frecuencia')
ax[1,0].grid()
# Diagrama de otras fuentes de energia
ax[1,1].hist(DataAll['Other sources'], bins=7, color='Black', alpha=0.7, label='Other sources', edgecolor='black')
ax[1,1].set_title('Generacion Electrica por source - Colombia')
ax[1,1].set_xlabel('Otras Fuentes de Generacion (GWh)')
ax[1,1].set_ylabel('Frecuencia')
ax[1,1].grid()
ax[1,2].axis('off')
plt.show()
```

DIAGRAMA DE PASTEL

```python
# Diagrama de torta con la distribucion de energia por fuente sumadas, asi se evita la sobrecarga de datos
Total_Energia_Otras = Total_Energia.iloc[[0, 2, 3, 4]].sum() 
Total_Energia_Otras
plt.pie([Total_Energia.iloc[1], Total_Energia_Otras], labels=['Hidro', 'Otras Fuentes'], autopct='%1.1f%%', startangle=50, 
        colors=['blue','pink'], explode=(0.05, 0.10), shadow=True, radius=1.2)
plt.title("Distribución de energía por fuente", fontsize=25, color='black', loc='center', pad=20)
plt.axis('equal')
plt.legend(loc='upper left', fontsize=10, title='Fuentes de energía', title_fontsize='13', shadow=True, facecolor='white', edgecolor='black')
plt.grid()
plt.show()
```

DIAGRAMA DE AREA

```python

# Variables
Años = EnergyCol['year']
Energias = EnergyCol[['Biofuels', 'Hydro', 'Solar PV', 'Wind', 'Other sources']]
colores = ['green', 'blue', 'orange', 'purple', 'gray']

# Gráfico de área apilada
plt.stackplot(Años, Energias.T, labels=Energias.columns, colors=colores, alpha=0.7)

# Estética
plt.title('Consumo de Energía Renovable en Colombia a lo Largo del Tiempo')
plt.xlabel('Años')
plt.ylabel('Generación Eléctrica (GWh)')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
