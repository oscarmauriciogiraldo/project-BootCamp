INICIO DE DIAGRAMAS IEA

```python
import numpy as np
import pandas as pd
```

```python
import os
file_path = 'Electricity generation by source - Colombia.csv'
if not os.path.exists(file_path):
	print(f"The file '{file_path}' does not exist in the current directory: {os.getcwd()}")
	print(f"Absolute path to check: {os.path.abspath(file_path)}")
	raise FileNotFoundError(f"The file '{file_path}' does not exist. Please ensure it is in the correct directory.")

Data = pd.read_csv(file_path)
Data.head()
```

```python
Data = Data.drop(columns=['Coal','Oil', 'Natural gas'])
```

IMPORTAR DIAGRAMAS

```python
Data['Country'] = 'Colombia'

Grupos = Data.groupby('Country')
EnergyCol = Grupos.get_group('Colombia')
EnergyCol
```

DIAGRAMAS

```python
import matplotlib.pyplot as plt
```

DIAGRAMA DE LINEAS

```python
DataAll = EnergyCol[['Biofuels','Hydro','Solar PV','Wind','Other sources']]
plt.plot(EnergyCol['year'], DataAll, label=DataAll.columns)
plt.title('Electricity generation by source - Colombia')
plt.xlabel('Años')
plt.ylabel('Electricity generation (GWh)')
plt.legend(DataAll.columns, loc='upper left')
plt.grid()
plt.show()
```

```python
Total_Energia = np.sum(EnergyCol.select_dtypes(include='number'))
Total_Energia = Total_Energia.iloc[1:]
Total_Energia
```

```python
Total_Energia
```

DIAGRAMA DE BARRAS

```python
plt.figure(figsize=(10, 5))
plt.bar(Total_Energia.index, Total_Energia.values , color=['aquamarine', 'green', 'grey', 'blue', 'orange'])
plt.title('Electricity generation by source - Colombia')
plt.xlabel('Años')
plt.ylabel('Electricity generation (GWh)')
plt.xticks(rotation=45)
```

```python
fig, ax = plt.subplots(2,3, figsize=(15,15), dpi=80, facecolor='w', edgecolor='k')
ax[0,0].hist(DataAll['Hydro'], bins=7, color='blue', alpha=0.7, label='Hydro', edgecolor='black')
ax[0,0].set_title('Electricity generation by source - Colombia')
ax[0,0].set_xlabel('Hydro generation (GWh)')
ax[0,0].set_ylabel('Frequency')
ax[0,0].grid()
ax[0,1].hist(DataAll['Biofuels'], bins=7, color='Green', alpha=0.7, label='Biofuels', edgecolor='black')
ax[0,1].set_title('Electricity generation by source - Colombia')
ax[0,1].set_xlabel('Biofuels generation (GWh)')
ax[0,1].set_ylabel('Frequency')
ax[0,1].grid()
ax[0,2].hist(DataAll['Solar PV'], bins=7, color='Yellow', alpha=0.7, label='Solar PV', edgecolor='black')
ax[0,2].set_title('Electricity generation by source - Colombia')
ax[0,2].set_xlabel('Solar PV generation (GWh)')
ax[0,2].set_ylabel('Frequency')
ax[0,2].grid()
ax[1,0].hist(DataAll['Wind'], bins=7, color='Grey', alpha=0.7, label='Wind', edgecolor='black')
ax[1,0].set_title('Electricity generation by source - Colombia')
ax[1,0].set_xlabel('Wind generation (GWh)')
ax[1,0].set_ylabel('Frequency')
ax[1,0].grid()
ax[1,1].hist(DataAll['Other sources'], bins=7, color='Black', alpha=0.7, label='Other sources', edgecolor='black')
ax[1,1].set_title('Electricity generation by source - Colombia')
ax[1,1].set_xlabel('Other sources generation (GWh)')
ax[1,1].set_ylabel('Frequency')
ax[1,1].grid()
ax[1,2].axis('off')
plt.show()
```

DIAGRAMA DE PASTEL

```python
Energias = ['Biofuels', 'Hydro', 'Solar PV', 'Wind', 'Other sources']
Filtered_Total_Energia = Total_Energia[Energias]  # Filter Total_Energia to match Energias
plt.pie(Filtered_Total_Energia, labels=Energias, autopct='%1.1f%%', startangle=50, colors=['pink', 'blue', 'yellow', 'grey', 'black'], explode=(0.05, 0.10, 0.20, 0.25, 0.30), shadow=True, radius=1.2)
plt.title("Distribución de energía por fuente", fontsize=25, color='black', loc='center', pad=20)
plt.axis('equal')
plt.legend(loc='upper left', fontsize=10, title='Fuentes de energía', title_fontsize='13', shadow=True, facecolor='white', edgecolor='black')
plt.grid()
plt.show()
```
